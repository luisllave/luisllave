```python
# Problem 3

locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    }, # <-- Note that Python does not complain about this extra comma. This is perfectly okay.
       #     The creators of Python saw that dictionaries tend to be changed a lot, so allowing this is a
       #     convenience for developers who need to add elements when needed.
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}


def eta(source,destination):   
    routes = list(legs.keys())
    if routes[0][0] == source:
        if routes[0][1] == destination:
            return(legs["upd","admu"]["travel_time_mins"])
        else:
            return((legs["upd","admu"]["travel_time_mins"])+(legs["admu","dlsu"]["travel_time_mins"]))
    if routes[1][0] == source:
        if routes[1][1] == destination:
            return(legs["admu","dlsu"]["travel_time_mins"])
        else:
            return((legs["admu","dlsu"]["travel_time_mins"])+(legs["dlsu","upd"]["travel_time_mins"]))
    if routes[2][0] == source:
        if routes[2][1] == destination:
            return(legs["dlsu","upd"]["travel_time_mins"])
        else:
            return((legs["dlsu","upd"]["travel_time_mins"])+(legs["upd","admu"]["travel_time_mins"]))
    
    
eta("dlsu","upd")
```
